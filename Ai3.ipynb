```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# membaca dataset
df = pd.read_csv('insurance.csv')

# mengambil fitur dan target
x = df['bmi'].values
y = df['charges'].values

# inisialisasi parameter
w = 0.0
b = 0.0
eta = 0.000001
epochs = 100

n = len(x)

# stochastic gradient descent
for epoch in range(epochs):
    for i in range(n):
        y_pred = w * x[i] + b

        # error sesuai PDF
        error = y_pred - y[i]

        # update parameter
        w = w - eta * error * x[i]
        b = b - eta * error

print("Weight =", w)
print("Bias =", b)

# prediksi semua data
y_pred_all = w * x + b

# hitung MSE
mse = np.mean((y_pred_all - y) ** 2)
print("MSE =", mse)

# visualisasi
plt.figure(figsize=(8,5))
plt.scatter(x, y)
plt.plot(np.sort(x), w*np.sort(x)+b)
plt.xlabel("BMI")
plt.ylabel("Charges")
plt.title("Linear Regression using SGD")
plt.show()
```
